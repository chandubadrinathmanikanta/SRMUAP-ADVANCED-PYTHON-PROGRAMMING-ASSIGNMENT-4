1)Write Python code to display the basic properties of the ca-netscience dataset.

Properties:

a)Number of nodes

b)Number of edges

c)List of nodes and edges

d)Degree of a node

e)Neighbours of a node

f)Connectivity of the graph

g)Diameter of a graph

h)Construct graph and visualize

```python
# Import necessary libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Function to load the graph from the file
def load_graph_from_file(file_path):
    G = nx.Graph()

    # Reading the graph from the input file
    with open(file_path, 'r') as f:
        for line in f:
            if line.strip():
                node1, node2 = map(int, line.strip().split())
                G.add_edge(node1, node2)

    return G

# Load the ca-netscience graph
file_path = 'ca-netscience.txt'  # Replace with the correct path to your file
G = load_graph_from_file(file_path)

# Check if the graph was loaded correctly
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("First 10 nodes:", list(G.nodes())[:10])
print("First 10 edges:", list(G.edges())[:10])
```

```python
# Function to display basic properties of the graph
def display_graph_properties(G, node_id):
    # Check if the node_id exists in the graph
    if node_id not in G:
        print(f"Node {node_id} does not exist in the graph.")
        return

    # Print basic properties of the graph
    print(f"Properties for Node {node_id}:")

    # 1. Number of nodes
    num_nodes = G.number_of_nodes()

    # 2. Number of edges
    num_edges = G.number_of_edges()

    # 3. List of nodes and edges
    nodes = list(G.nodes())
    edges = list(G.edges())

    print("Number of nodes:", num_nodes)
    print("Number of edges:", num_edges)
    print("First 10 nodes:", nodes[:10], "...")  # Displaying the first 10 nodes for brevity
    print("First 10 edges:", edges[:10], "...")  # Displaying the first 10 edges for brevity

    # 4. Degree of a specific node
    node_degree = G.degree(node_id)
    print(f"Degree of node {node_id}: {node_degree}")

    # 5. Neighbors of a specific node
    node_neighbors = list(G.neighbors(node_id))
    print(f"Neighbors of node {node_id}: {node_neighbors}")

    # 6. Connectivity of the graph (check if the graph is connected)
    is_connected = nx.is_connected(G)
    print(f"Is graph connected?: {is_connected}")

    # 7. Diameter of the graph (only if connected)
    if is_connected:
        diameter = nx.diameter(G)
        print(f"Diameter of the graph: {diameter}")
    else:
        print("Graph is not connected, so diameter is not defined.")

# Call the function to display graph properties for a specific node (e.g., node 0)
display_graph_properties(G, node_id=0)
```

```python
# Visualization of the graph
plt.figure(figsize=(10, 10))
nx.draw(G, with_labels=True, node_size=500, node_color="skyblue", font_size=10, font_color="black")
plt.title("Visualization of the ca-netscience graph")
plt.show()
```

2)Write Python code to find out the top 10 nodes for the ca-netscience dataset
using the following centrality measures.

a)Degree Centrality

b)Closeness Centrality

c)Betweenness Centrality

d)Harmonic Centrality

e)Page Rank Centrality

```python
# Import necessary libraries
import networkx as nx
import matplotlib.pyplot as plt

# Function to load the graph from the file
def load_graph_from_file(file_path):
    G = nx.Graph()

    # Reading the graph from the input file
    with open(file_path, 'r') as f:
        for line in f:
            if line.strip():
                node1, node2 = map(int, line.strip().split())
                G.add_edge(node1, node2)

    return G

# Load the ca-netscience graph
file_path = 'ca-netscience.txt'  # Replace with the correct path to your file
G = load_graph_from_file(file_path)
```

```python
# Function to calculate top 10 nodes based on different centrality measures
def calculate_top_10_centrality_measures(G):
    # a) Degree Centrality
    degree_centrality = nx.degree_centrality(G)
    top_10_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

    # b) Closeness Centrality
    closeness_centrality = nx.closeness_centrality(G)
    top_10_closeness = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

    # c) Betweenness Centrality
    betweenness_centrality = nx.betweenness_centrality(G)
    top_10_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

    # d) Harmonic Centrality
    harmonic_centrality = nx.harmonic_centrality(G)
    top_10_harmonic = sorted(harmonic_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

    # e) Page Rank Centrality
    page_rank_centrality = nx.pagerank(G)
    top_10_page_rank = sorted(page_rank_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

    # Display top 10 for each centrality measure
    print("\nTop 10 Nodes by Degree Centrality:")
    for node, centrality in top_10_degree:
        print(f"Node {node}: {centrality}")

    print("\nTop 10 Nodes by Closeness Centrality:")
    for node, centrality in top_10_closeness:
        print(f"Node {node}: {centrality}")

    print("\nTop 10 Nodes by Betweenness Centrality:")
    for node, centrality in top_10_betweenness:
        print(f"Node {node}: {centrality}")

    print("\nTop 10 Nodes by Harmonic Centrality:")
    for node, centrality in top_10_harmonic:
        print(f"Node {node}: {centrality}")

    print("\nTop 10 Nodes by Page Rank Centrality:")
    for node, centrality in top_10_page_rank:
        print(f"Node {node}: {centrality}")

# Call the function to calculate and display top 10 centrality measures
calculate_top_10_centrality_measures(G)
```

```python
# Visualization of top 10 nodes based on Degree Centrality
top_10_degree_nodes = [node for node, _ in sorted(nx.degree_centrality(G).items(), key=lambda x: x[1], reverse=True)[:10]]

# Draw the graph with highlighted top 10 nodes by Degree Centrality
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_size=300, node_color="skyblue", font_size=8)
nx.draw_networkx_nodes(G, pos, nodelist=top_10_degree_nodes, node_color="red", node_size=500, label="Top 10 Degree Centrality")
plt.title("Top 10 Nodes by Degree Centrality (Highlighted in Red)")
plt.legend()
plt.show()
```
